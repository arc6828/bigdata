```python
df = pd.read_csv('8bits.csv')
```

```python
tree_map_dict = {};
for index, row in df.iterrows():
    #print(row['eight_bits'], row['volume'])
    
    key_root = "root"
    key8 = str(row['eight_bits']);
    key6 = str(row['eight_bits'])[:6];
    key4 = str(row['eight_bits'])[:4];
    volume = row['volume']
    
    #KEY8
    if key8 not in tree_map_dict :         
        #CREATE IF NOT EXIST        
        tree_map_dict[key8] = {"location":key8 , "parent" : key6 , "volume" : volume} 
    else: 
        #UPDATE VOLUME IF EXIST
        tree_map_dict[key8]["volume"] += volume
    
    #KEY6
    if key6 not in tree_map_dict :         
        #CREATE IF NOT EXIST        
        tree_map_dict[key6] = {"location":key6 , "parent" : key4 , "volume" : volume} 
    else: 
        #UPDATE VOLUME IF EXIST
        tree_map_dict[key6]["volume"] += volume
    
    #KEY4
    if key4 not in tree_map_dict :         
        #CREATE IF NOT EXIST        
        tree_map_dict[key4] = {"location":key4 , "parent" : key_root , "volume" : volume} 
    else: 
        #UPDATE VOLUME IF EXIST
        tree_map_dict[key4]["volume"] += volume
    print(key_root)
        
        
    #KEY ROOT
    
    if key_root not in tree_map_dict :         
        #CREATE IF NOT EXIST        
        tree_map_dict[key_root] = {"location":key_root , "parent" : null , "volume" : volume} 
    else: 
        #UPDATE VOLUME IF EXIST
        tree_map_dict[key_root]["volume"] += volume
    
```
